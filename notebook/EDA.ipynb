```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
train_data = pd.read_csv("./data/ObesityDataSet.csv")
display(train_data.head())
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

# Specify columns to be encoded
columns_to_be_encoded = ['CAEC', 'CALC', 'MTRANS']
encoded = encoder.fit_transform(train_data[columns_to_be_encoded])
encoded_array = encoded.toarray()
encoded_feature_names = encoder.get_feature_names_out(columns_to_be_encoded)

encoded_df = pd.DataFrame(encoded_array, columns=encoded_feature_names)
train_data_encoded = train_data.drop(columns=columns_to_be_encoded)
train_data_encoded = pd.concat([train_data_encoded, encoded_df], axis=1)
print(train_data_encoded.head())
```

```python
gender_mapping = {'Male':0, 'Female':1}
train_data_encoded['Gender'] = train_data_encoded['Gender'].map(gender_mapping)
family_mapping = {'yes':1, 'no':0}
train_data_encoded['family_history_with_overweight'] = train_data_encoded['family_history_with_overweight'].map(family_mapping)
FAVC_mapping = {'yes':1, 'no':0}
train_data_encoded['FAVC'] = train_data_encoded['FAVC'].map(FAVC_mapping)
SCC_mapping = {'yes':1, 'no':0}
train_data_encoded['SCC'] = train_data_encoded['SCC'].map(SCC_mapping)
train_data_encoded.drop(columns = 'SMOKE', inplace = True)
```

```python
label = train_data_encoded['NObeyesdad']
train_data_encoded.drop(columns = 'NObeyesdad', inplace=True)
```
